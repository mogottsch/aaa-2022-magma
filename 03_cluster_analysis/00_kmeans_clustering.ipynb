## Hard Clustering: K-means++

Now we start with clustering based on all features selected.

```python
X_cluster_features = X_scaled_df.drop(columns=["land_use_start", "land_use_end"])
```

```python
# defining the maximum quantity of centers
k_max = 10

clusters = []
losses = []

# train each model and save losses
for k in range(k_max):
    model = KMeans(n_clusters=k + 1, random_state=4711)
    model.fit(X_cluster_features)
    clusters.append(k + 1)
    losses.append(model.inertia_)
```

```python
# plot losses
plt.plot(clusters, losses)
plt.show()
```

We select a number of 5 clustered based on the loss curve. 

```python
# run again with k=5
five_means = KMeans(n_clusters=5, random_state=4711)
five_means.fit(X_cluster_features)
five_means.predict(X_cluster_features)
```

```python
# map cluster to data
X["cluster"] = five_means.labels_
```

```python
X_cluster_means = X.groupby("cluster").mean()
X_cluster_means.iloc[:,0:13]
```

```python
# function for plotting violin plots to compare clusters regarding different features
def show_cluster_values(columns):
    fig, axs = plt.subplots(int(len(columns) / 2), 2)
    fig.set_size_inches(10, 10)
    sns.set(font_scale = 1.25)
    plots_row = 0
    plots_col = 0
    for col in columns:
        sns.violinplot(
            x=X["cluster"],
            y=X[col],
            data=X,
            palette="magma",
            ax=axs[plots_row][plots_col]
        )
        axs[plots_row][plots_col].set_xlabel("Cluster")
        axs[plots_row][plots_col].set_ylabel("".join(l.capitalize() + " " for l in col.replace("_"," ").split()))
        if plots_col == 1:
            plots_row = plots_row + 1
            plots_col = 0
        else:
            plots_col = 1
    plt.tight_layout()
    plt.savefig("../00_data/processed/clusters_" + "".join(columns[0]) + ".png")
```

```python
# function for plotting scatter plots to compare clusters regarding different features
def show_cluster_values_jitter(columns):
    fig, axs = plt.subplots(int(len(columns) / 2), 2)
    fig.set_size_inches(10, 10)
    sns.set(font_scale = 1.25)
    plots_row = 0
    plots_col = 0
    for col in columns:
        sns.stripplot(
            x=X["cluster"],
            y=X[col],
            data=X,
            jitter=True,
            palette="magma",
            ax=axs[plots_row][plots_col]
        )
        axs[plots_row][plots_col].set_xlabel("Cluster")
        axs[plots_row][plots_col].set_ylabel("".join(l.capitalize() + " " for l in col.replace("_"," ").split()))
        if plots_col == 1:
            plots_row = plots_row + 1
            plots_col = 0
        else:
            plots_col = 1
    plt.tight_layout()
    plt.savefig("../00_data/processed/clusters_" + "".join(columns[0]) + ".png")
```

## K-means++ partial clustering

In this section we cluster using only a subset of the features and then take a look at all features per cluster. Thus, we hope to identify how features interact and compliment each other. Because we do not discuss this part in the paper, not only plots have been analyzed and interpreted in detail. However, there are some interesting correlations to be discovered.

First, we define some useful functions.

```python
# defining the maximum quantity of centers
def find_k(X_scaled,k_max):

    clusters = []
    losses = []

    # train each model and save losses
    for k in range(k_max):
        model = KMeans(n_clusters=k + 1, random_state=4711)
        model.fit(X_scaled)
        clusters.append(k + 1)
        losses.append(model.inertia_)

    plt.plot(clusters, losses)
    plt.show()
```

```python

def cluster_plots(columns, x1, x2, cluster_column):
    fig, axs = plt.subplots(x1, x2)
    fig.set_size_inches(10, 10)
    plots_row = 0
    plots_col = 0
    for col in columns:
        sns.violinplot(
            x=X[cluster_column],
            y=X[col],
            data=X,
            palette="magma",
            ax=axs[plots_row][plots_col]
        )
        axs[plots_row][plots_col].set_xlabel("Cluster")
        axs[plots_row][plots_col].set_ylabel("".join(l.capitalize() + " " for l in col.replace("_"," ").split()))
        if plots_col == 1:
            plots_row = plots_row + 1
            plots_col = 0
        else:
            plots_col = 1
    plt.tight_layout()
```

```python
def run_kmeans(X_scaled, n, name):
    k_means = KMeans(n_clusters=n, random_state=4711)
    k_means.fit(X_scaled)
    k_means.predict(X_scaled)

    X[name] = k_means.labels_
```
